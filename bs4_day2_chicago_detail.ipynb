```python
price_list = []
adress_list = []

for idx in df.index :
    req = Request(url=df['URL'][idx], headers={'User-Agent':generate_user_agent()})
    page = urlopen(req)
    soup = BeautifulSoup(page, "html.parser")
    
    text = soup.find(class_="addy").get_text() 
    need_data = re.split(", ",text)[0]
    price = re.search("\$\d+.(\d+)?", need_data).group()
    adress = need_data[len(price)+1:]
    
    price_list.append(price)
    adress_list.append(adress)
    
price_list, adress_list
```

```python
df['Price'] = price_list
df['Address'] = adress_list
```

```python
a = 12

ep_a = driver.find_element(By.CLASS_NAME, "list.detail.eplist") # a 가 변수이므로
ep_info_list = ep_a.find_elements(By.CLASS_NAME, "info")
ep_name_list = ep_a.find_elements(By.TAG_NAME, "strong")
ep_name_rating_list = ep_a.find_elements(By.CLASS_NAME, "ipl-rating-star.small")
ep_date_list = ep_a.find_elements(By.CLASS_NAME, "airdate")


data = []

for i in range(len(ep_info_list)): #len 하여 info 갯수 24개를 Range(24) 하여 i에 넣어준다. #에피소드이름, 제목, 날짜, 평점을 넣어야 하므로 info가 제일 나을 것이다.
    ep_name = ep_name_list[i].text
    ep_date = ep_date_list[i].text
    ep_rating = ep_name_rating_list[i].text
    
    data.append([f"{a}.{i+1}", ep_name, ep_date, ep_rating])#{i+1} 해야 1부터 시작된다.

df_12 = pd.DataFrame(data, columns=["Episode Number", "Name", "Date", "Rating"])

df_12.index = range(1,25) #0 부터 출력되므로 1부터 나오게 범위 지정해주기


df_12
```

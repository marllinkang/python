## 01. 기본 그래프

```python
datas = [1, 2, 3, 4, 7, 4, 8, 9, 2, 3]
```

```python
plt.figure(figsize=(3, 3))  
plt.plot(datas)
plt.show()
```

```python
x_datas = np.arange(0, 200, 20)
y_datas = np.random.randint(10, size=10)
```

```python
len(x_datas), len(y_datas)
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(x_datas, y_datas)

plt.title('점심은 뭐 먹지?')
plt.xlabel('구내식당?')
plt.ylabel('햄버거?')

plt.show();
```

## 02. 스타일

```python
plt.figure(figsize=(3, 3))
plt.plot(y_datas, 'r^-.') #r빨 b파
plt.show();
```

```python
plt.figure(figsize=(3, 3))
plt.plot(y_datas, c='g', lw=5, ls=':', marker='*', mec='r', mew=1, mfc='y') #마커 별 c그린 lw선의 굵기
plt.show();
```

## 03. 축 설정

```python
plt.figure(figsize=(5, 2))
plt.plot(y_datas)

plt.xlim(-2, 11)
plt.ylim(-2, 15)

plt.show();
```

```python
plt.figure(figsize=(5, 3))
plt.plot(y_datas, 'r^-.')

plt.xticks(range(0, 10))
plt.yticks(range(0, 11, 5))

plt.show();
```

## 04. 그래프 여러 개 그리기 

```python
first = np.random.randint(10, size=5)
second = np.random.randint(10, size=5)
third = np.random.randint(10, size=5)
```

```python
plt.figure(figsize=(3, 3))

plt.plot(first)
plt.plot(second)
plt.plot(third)

plt.show();
```

```python
datas = [] 

for _ in range(3):
    datas.append(np.random.randint(10, size=10))

datas
```

```python
datas = [np.random.randint(10, size=10) for _ in range(3)]
datas
```

```python
plt.figure(figsize=(5, 5))
plt.plot(datas[0], 'ro--', datas[1], 'b*--', datas[2], 'g^-')
# plt.plot(datas[1], 'b*--')
plt.grid()
plt.show();
```

```python
plt.figure(figsize=(5, 5))

plt.plot(datas[0], label='data1')
plt.plot(datas[1], label='data2')
plt.plot(datas[2], label='data3')

plt.legend(loc='upper right')
plt.grid()
plt.show();
```

```python
# 10개 길이
datas = [np.random.randint(10, size=10) for _ in range(10)]
len(datas)
```

```python
plt.figure(figsize=(5, 5))

for idx in range(10):
    plt.plot(datas[idx], label=f'data{idx}')

plt.legend(loc='upper right')
plt.show();
```

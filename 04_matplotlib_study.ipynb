## 01. 기본 그래프

```python
datas = [1, 2, 3, 4, 7, 4, 8, 9, 2, 3]
```

```python
# 그럎르를 그리려면 밑바탕 도화지, 그림, 보여줌이 필요함
# 그래프를 그리려면 3가지가 기본요소

plt.figure(figsize=(3,3)) # 도화지
plt.plot(datas) # 그림
plt.show() # 보여줌
```

```python
# x와 y 지정하기위해서는 둘의 길이가 같아야한다.
x_datas = np.arange(0,200,20)
y_datas = np.random.randint(10,size=10)
#len(x_datas,y_datas)
```

```python
plt.figure(figsize=(5,5))  # 도화지
plt.plot(x_datas,y_datas)  # 그림
plt.show()  # 보여주기
```

```python
plt.figure(figsize=(5,5))
plt.plot(x_datas,y_datas)
plt.title('점심 뭐먹지?')
plt.xlabel('구내식당?')
plt.ylabel('햄버거?')
plt.show();
```
